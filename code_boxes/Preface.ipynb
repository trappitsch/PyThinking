## Py code 0.4

R comes with a default dataset called cars. More information to be found by typing `?cars` in R. There are however standard datasets available in scikit-learn, more information can be found [here](https://scikit-learn.org/stable/datasets/index.html). Here we will use the `diabetes` dataset. The code in the book here is simply an example ofs ome mroe complex code and shows what R can do, however, every snipped should be translated if somehow possible.


```python
from sklearn import datasets, linear_model
import numpy as np
import matplotlib.pyplot as plt

X, y = datasets.load_diabetes(return_X_y=True)

# y: quantitative measure of disease progression one year after baseline
# X: various predictive values
# X[:, 2]: body mass index -> what we load
xcrim = X[:, np.newaxis, 2]

# create the linear model for xcrim and y
regr = linear_model.LinearRegression()
regr.fit(xcrim, y)

# print the coefficients
print(regr.coef_)

# to plot the residuals we need to first calculate them
# calculate the predictions of the model for the test data
y_pred = regr.predict(xcrim)

residuals = y - y_pred

# use matplotlib to plot the residuals
plt.plot(xcrim, residuals, 'o')
plt.show()

plt.figure(2)
plt.plot(xcrim, y, 'o')
plt.plot(xcrim, y_pred, '-')
```
